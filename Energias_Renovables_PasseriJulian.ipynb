```python
#Seccion 9: Elemino ciertas columnas para realizar un modelo predictivo, tambien limpio datos vacios para evitar NaN
df_cleaned = df.dropna(subset=['Value_co2_emissions_kt_by_country'])

X = df_cleaned.drop(columns=['Entity'], axis=1) #Axis indica columnas, hago un drop de Entity debido a errores de tipos de datos
y =df_cleaned.Value_co2_emissions_kt_by_country   #Defino el target de estudio, emisiones de CO2 por Pais

from sklearn.model_selection import train_test_split #Libereria para separar entre train y test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#Seccion 10: Elemino ciertas columnas para realizar un modelo predictivo 
#Implemento otro modelo, debido al tipo de datos manejados en CO2 (valores continuos)
from sklearn.tree import DecisionTreeRegressor 
# Lo entrenamos
tree_regressor = DecisionTreeRegressor()
tree_regressor.fit(X_train, y_train)
```

```python
#Seccion 11: Busco elementos vacios de datos
missing_values = y_train.isnull().sum()
print("Valores faltantes para entrenar en y_train:", missing_values)
```

```python
#Seccion 12: Realizo las predicciones
y_pred = tree_regressor.predict(X_test)
```

```python
#Seccion 13: Evaluamos al modelo
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#Errores medios absolutos (MAE), errores medios cuadrados (MSE) y coeficiente de determinaciÃ³n (R2) en el contexto de la modelizaciÃ³n de regresiÃ³n:

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
#Seccion 14: Creamos Grafica

metrics = ['MAE', 'MSE', 'R-squared']
values = [mae, mse, r2]
log_values = np.log10(values) #Aplico una escala logaritmica 

# Crear un grÃ¡fico de barras con valores escalados en escala logarÃ­tmicaplt.figure(figsize=(8, 5))
plt.bar(metrics, log_values, color=['blue', 'green', 'orange'])
plt.xlabel('Metrics')
plt.ylabel('Logarithm of Values (log10)')
plt.title('Logarithmic Scale Model Evaluation Metrics')

# Mostrar los valores reales en la parte superior de las barras
for i, v in enumerate(log_values):
    plt.text(i, v + 0.02, str(round(values[i], 2)), ha='center', va='bottom')

plt.show()
```
